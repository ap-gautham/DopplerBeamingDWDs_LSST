# Analyze the Generated Population

There are 10 million SeBa DWD

There are 100 million DWD coordinates

----
1. First sample (with replacement) the SeBa DWDs and geneate a sample of 100 million DWDs with coordinates
    - Take the coordinates and time, then sample the SeBa DWDs formed at that time.
        - It is probably best to first group the SeBa observations by time, so that scaling to 100 million is efficient.
    - Take the LSST observation baseplan and use to check which coordinates should be assigned and which shouldn't. This will save time AND space.
    - __If a coordinate is assigned long-period binary (P>30 hours), then simply discard it. This will save space with no cost since these binaries will not show any Doppler beaming.__
    
----

2. Then use the Bergeron cooling models to generate photometry for the DWDs
    - If the WD parameters are outside the Bergeron grid, then discard them as they are not yet on the WD cooling track.
    - Assign inclination
    - Assign only DWDs with period < 30 hours. This will save a significant amount of time.
        - You will need to assign long period DWDs coordinates because they likely correlate with the formation time. However, don't waste time assigning photometry and instead set them all to zero.
        - Better yet, if a coordinate is assigned long-period binary them simply discard it. This will save space as well.
 
----   
#### In all the above, remember to balance time and space. It is best to write to a h5 file periodically. But writing often will be hard to parallelize.


## With these steps I will have a simulated sample of DWDs. I can then generate lightcurves and recover the period.

# Package

```python
import numpy as np
import astropy
from astropy import units
from astropy import *
import time
from tqdm import tqdm
import matplotlib.pyplot as plt
import glob
import os
#from multiprocess import Pool
#import multiprocess

import h5py
    
# SEBA output columns

seba_cols = {'binary_ID': 0,\
'binary_Type': 1,\
'time': 2,\
'a_Rsun': 3,\
'star_ID1': 4,\
'star_Type1': 5,\
'M1': 6,\
'R1': 7,\
'logTeff1': 8,\
'star_ID2': 9,\
'star_Type2': 10,\
'M2': 11,\
'R2': 12,\
'logTeff2': 13}

# Binary Types

seba_binary_types = {2:'detached',\
3:'semi detached + stable mass transfer', \
4:'contact',\
5:'CE (gamma)',\
6:'double spiral-in',\
7:'merged',\
8:'disrupted',\
9:'CE (alpha)'}

# Mass Transfer Types

seba_MT_types = {1:'on nuclear time scale',\
2:'on angular momentum loss timescale',\
3:'on thermal time scale',\
4:'CE due to dynamics',\
5:'CE due to Darwin Riemann instability'}

# Stellar Types

seba_star_types = {1:'planet',\
2:'brown dwarf',\
3:'main sequence',\
5:'hertzsprung gap',\
6:'sub-giant',\
7:'core helium burning star',\
8:'agb',\
10:'helium star',\
11:'helium giant',\
12:'carbon-oxygen white dwarf',\
13:'helium white dwarf',\
14:'oxygen-neon white dwarf',\
18:'neutron star',\
19:'black hole',\
20:'disintegrated'}

global simuls

def f_separationToPeriod(a,m1,m2):
    return (2*np.pi*(a**3/astropy.constants.G/(m1+m2))**0.5).decompose()

def old_f_mass_radius_to_logg(M,R): 
    G = astropy.constants.G
    return np.log10((G*M/(R)**2).to(units.centimeter/units.second**2).value)

def f_mass_radius_to_logg(M_msun,R_rsun): 
    return np.log10((M_msun/R_rsun**2)) + 4.438

i_grid = np.arange(0,90,0.01)
pdf_i = lambda i: np.sin(i*np.pi/180)
prob_i = pdf_i(i_grid)/np.sum(pdf_i(i_grid))
f_sample_i = lambda size=1: np.random.choice(i_grid,p=prob_i,size=size)
```

```python
from dustmaps.sfd import SFDQuery
from astropy.coordinates import SkyCoord

sfd_EBV = SFDQuery()

Rv = 3.1
sfd_Av = lambda coords: sfd_EBV(coords)*Rv

def f_Av(coords,d_kpc):
    l = coords.galactic.l
    b = coords.galactic.b
    
    abs_b = np.maximum(np.abs(b.value),1)
    
    zs = 0.3#*units.kiloparsec
    return sfd_Av(coords)*(1-np.exp(-d_kpc*np.sin(abs_b*np.pi/180)/zs))

bandbass_EBVconversion_dict = {'SDSS_u':4.239,\
                       'PanSTARRS_g':3.172,\
                       'PanSTARRS_r':2.271,\
                       'PanSTARRS_i':1.682,\
                       'PanSTARRS_z':1.322,\
                       'PanSTARRS_y':1.087}

bandbass_Abconversion_dict = {'SDSS_u':1.37,\
                       'PanSTARRS_g':1.02,\
                       'PanSTARRS_r':0.73,\
                       'PanSTARRS_i':0.54,\
                       'PanSTARRS_z':0.43,\
                       'PanSTARRS_y':0.35}

bandbass_Abconversion = np.array(list(bandbass_Abconversion_dict.values()))
```

```python
import numpy as np
import scipy
from astropy.table import Table
from astropy.io import ascii

header = 'Teff,logg,M/Mo,Mbol,BC,'\
            +'JKC_U,JKC_B,JKC_V,JKC_R,JKC_I,'\
            +'2MASS_J,2MASS_H,2MASS_Ks,'\
            +'MKO_Y,MKO_J,MKO_H,MKO_K,'\
            +'WISE_W1,WISE_W2,WISE_W3,WISE_W4,'\
            +'IRAC_S3.6,IRAC_S4.5,IRAC_S5.8,IRAC_S8.0,'\
            +'SDSS_u,SDSS_g,SDSS_r,SDSS_i,SDSS_z,'\
            +'PanSTARRS_g,PanSTARRS_r,PanSTARRS_i,PanSTARRS_z,PanSTARRS_y,'\
            +'GAIA_G2,GAIA_G2_BP,GAIA_G2_RP,'\
            +'GAIA_G3,GAIA_G3_BP,GAIA_G3_RP,'\
            +'GALEX_FUV,GALEX_NUV,'\
            +'Age'
header = header.split(',')

bergeron_DA = ascii.read('../02_Data/Bergeron_DA',data_start=1)
bergeron_DA.rename_columns(bergeron_DA.colnames,header)

logg, teff = np.unique(list(bergeron_DA['logg'])), np.unique(list(bergeron_DA['Teff']))

data = []
for ii in range(len(logg)):
    temp = []
    for jj in range(len(teff)):
        mask = (bergeron_DA['logg']==logg[ii])*(bergeron_DA['Teff']==teff[jj])
        temp.append(list(bergeron_DA[mask]['SDSS_u','PanSTARRS_g','PanSTARRS_r','PanSTARRS_i','PanSTARRS_z','PanSTARRS_y'][0]))
    data.append(temp)

bergeron_DA_interpolator = scipy.interpolate.RegularGridInterpolator((logg,teff),data)
```

```python
# All coordinates

input_array = np.load('01_results/input_array.npy')
input_array = input_array[np.argsort(input_array['age'])]

# Unique time 10 million stars

input_unique_array = np.load('01_results/input_unique_array.npy')
```

```python
# SeBa DWDs

data_directory = '/home/gadaman1/data-znadia1/01_Projects/01_WD/07_LSST_DWD'
h5_file_handle = h5py.File(data_directory+f'/SeBa_CurrentGalacticBinaries.h5', 'r')
dataset = h5_file_handle[f'/SeBa_CurrentGalacticBinaries']   

# Get the unique times and indices
SeBa_unique_time,SeBa_unique_args,SeBa_unique_count = (np.unique(dataset['time'],\
                                                                 return_index=True,return_counts=True))

# Combine times,indicies,counts so that you can fetch all DWDs created at a fixed time
SeBa_unique_time_arg_count = dict(zip(SeBa_unique_time,zip(SeBa_unique_args,SeBa_unique_count)))
```

```python
# Combine times,indicies,counts so that you can fetch all stars created at a fixed time
full_unique_time,full_unique_args,full_unique_count = (np.unique(input_array['age'],\
                                                                 return_index=True,return_counts=True))
full_unique_time_arg_count = dict(zip(full_unique_time,zip(full_unique_args,full_unique_count)))

# Combine times,indicies,counts so that you can fetch all unique 10 million stars created at a fixed time
simul_unique_time_count = dict(zip(input_unique_array['age_unique'],input_unique_array['age_unique_counts']))
```

```python
# Combine coordinates and DWD parameters

input_field_names = ['age',\
                     'r_cylindrical', 'phi_cylindrical', 'z_cylindrical',\
                     'RA_ICRS','DEC_ICRS', 'd_ICRS']
input_field_dtype = [np.int32,\
                     np.float32,np.float32,np.float32,\
                     np.float32,np.float32,np.float32]

SeBa_field_names = ['a_Rsun',\
                      'star_Type1','M1','R1','logTeff1',\
                      'star_Type2','M2','R2','logTeff2',]
SeBa_field_dtype = [np.float32,\
                      np.int16,np.float32,np.float32,np.float32,\
                      np.int16,np.float32,np.float32,np.float32,]

combined_array = dict(zip(['age',\
                           'r_cylindrical', 'phi_cylindrical', 'z_cylindrical',\
                           'RA_ICRS','DEC_ICRS', 'd_ICRS',\
                           'a_Rsun',\
                           'star_Type1','M1','R1','logTeff1',\
                           'star_Type2','M2','R2','logTeff2',\
                           'period_hour','inclination','Av',\
                           'u1','g1','r1','i1','z1','y1',\
                           'u2','g2','r2','i2','z2','y2',\
                           'u','g','r','i','z','y'],\
                            [[],\
                            [],[],[],\
                            [],[],[],\
                            [],\
                            [],[],[],[],\
                            [],[],[],[],\
                            [],[],[],\
                            [],[],[],[],[],[],\
                            [],[],[],[],[],[],\
                            [],[],[],[],[],[]]))

combined_field = ['age',\
                   'r_cylindrical', 'phi_cylindrical', 'z_cylindrical',\
                   'RA_ICRS','DEC_ICRS', 'd_ICRS',\
                   'a_Rsun',\
                   'star_Type1','M1','R1','logTeff1',\
                   'star_Type2','M2','R2','logTeff2',\
                   'period_hour','inclination','Av',\
                   'u1','g1','r1','i1','z1','y1',\
                   'u2','g2','r2','i2','z2','y2',\
                   'u','g','r','i','z','y']
combined_dtype = [np.int32,\
                  np.float32,np.float32,np.float32,\
                  np.float32,np.float32,np.float32,\
                  np.float32,\
                  np.int16,np.float32,np.float32,np.float32,\
                  np.int16,np.float32,np.float32,np.float32,\
                  np.float32,np.float32,np.float32,\
                  np.float32,np.float32,np.float32,np.float32,np.float32,np.float32,\
                  np.float32,np.float32,np.float32,np.float32,np.float32,np.float32,\
                  np.float32,np.float32,np.float32,np.float32,np.float32,np.float32]

for ii in tqdm(range(len(full_unique_time))):
    # Select all stars formed at time full_unique_time[ii]
    temp_full_arg, temp_full_count = full_unique_time_arg_count[full_unique_time[ii]]
    
    if full_unique_time[ii] in SeBa_unique_time:
        # If some stars formed DWD then add it to our list
        
        # Select SeBa DWD output
        temp_SeBa = dataset[SeBa_unique_time_arg_count[full_unique_time[ii]][0]:\
                             SeBa_unique_time_arg_count[full_unique_time[ii]][0]+\
                                SeBa_unique_time_arg_count[full_unique_time[ii]][1]]
        
        # Select star coordinates
        temp_input = input_array[temp_full_arg:\
                                 temp_full_arg+temp_full_count]
        
        # Count the number of WDs formed from n_simul number of simulated stars
        # Calculate the probability of WD formed
        n_wd = SeBa_unique_time_arg_count[full_unique_time[ii]][1]
        n_simul = simul_unique_time_count[full_unique_time[ii]]
        p_wd = n_wd/n_simul   
        
        # Flag which of the total coordinates can become a WD and assign WD parameters at the coordinate
        wd_flag = np.random.choice([True,False],p=[p_wd,1-p_wd],size=temp_full_count)
        temp_input_wd = temp_input[np.argwhere(wd_flag)[:,0]]        
        temp_SeBa = np.random.choice(temp_SeBa,len(temp_input_wd))
        
        t_DWD_m1 = temp_SeBa['M1']*units.M_sun
        t_DWD_m2 = temp_SeBa['M2']*units.M_sun
        t_DWD_a = temp_SeBa['a_Rsun']*units.R_sun
        t_DWD_period = f_separationToPeriod(t_DWD_a,t_DWD_m1,t_DWD_m2).to(units.hour)
        
        t_shortPeriodDWDs = np.argwhere((t_DWD_period<30*units.hour)*(temp_input_wd['d_ICRS']<5))[:,0]
        
        if len(t_shortPeriodDWDs) > 0 :
            M1 = temp_SeBa[t_shortPeriodDWDs]['M1']
            R1 = temp_SeBa[t_shortPeriodDWDs]['R1']
            logg1 = f_mass_radius_to_logg(M1,R1)
            Teff1 = 10**temp_SeBa[t_shortPeriodDWDs]['logTeff1']

            M2 = temp_SeBa[t_shortPeriodDWDs]['M2']
            R2 = temp_SeBa[t_shortPeriodDWDs]['R2']
            logg2 = f_mass_radius_to_logg(M2,R2)
            Teff2 = 10**temp_SeBa[t_shortPeriodDWDs]['logTeff2']
                
            inclincations = f_sample_i(size=len(t_shortPeriodDWDs))
            
            # Star's coordinate
            RA = temp_input_wd[t_shortPeriodDWDs]['RA_ICRS']*units.degree
            DEC = temp_input_wd[t_shortPeriodDWDs]['DEC_ICRS']*units.degree
            coord = SkyCoord(RA,DEC,frame='icrs')

            d_kpc = temp_input_wd[t_shortPeriodDWDs]['d_ICRS']

            # Dust extinction
            Av = f_Av(coord,d_kpc)
                    
            for ii in range(len(t_shortPeriodDWDs)):
                if (7 < logg1[ii] and logg1[ii] < 9) and (7 < logg2[ii] and logg2[ii] < 9) and \
                    (1500 < Teff1[ii] and Teff1[ii] < 150000) and (1500 < Teff2[ii] and Teff2[ii] < 150000):
                                    
                    # Observed magnitude of star 1
                    abs_mags = bergeron_DA_interpolator((logg1[ii],Teff1[ii]))
                    obs_mags1 = abs_mags + 5*np.log10(d_kpc[ii]) + 10 + bandbass_Abconversion*Av[ii]
                    u1,g1,r1,i1,z1,y1 = obs_mags1

                    # Observed magnitude of star 2
                    abs_mags = bergeron_DA_interpolator((logg2[ii],Teff2[ii]))
                    obs_mags2 = abs_mags + 5*np.log10(d_kpc[ii]) + 10 + bandbass_Abconversion*Av[ii]
                    u2,g2,r2,i2,z2,y2 = obs_mags2

                    # Observed magnitude of both stars
                    obs_mags = -2.5*np.log10(10**(-0.4*obs_mags1)+10**(-0.4*obs_mags2))
                    u,g,r,i,z,y = obs_mags
                
                    if r < 24:
                        for field in input_field_names:
                            combined_array[field].append(temp_input_wd[t_shortPeriodDWDs[ii]][field])
                        for field in SeBa_field_names:
                            combined_array[field].append(temp_SeBa[t_shortPeriodDWDs[ii]][field])

                        # Orbital Period
                        combined_array['period_hour'].append(t_DWD_period[t_shortPeriodDWDs[ii]].value)
                        combined_array['inclination'].append(inclincations[ii])

                        combined_array['Av'].append(Av[ii])

                        combined_array['u1'].append(u1)
                        combined_array['g1'].append(g1)
                        combined_array['r1'].append(r1)
                        combined_array['i1'].append(i1)
                        combined_array['z1'].append(z1)
                        combined_array['y1'].append(y1)

                        combined_array['u2'].append(u2)
                        combined_array['g2'].append(g2)
                        combined_array['r2'].append(r2)
                        combined_array['i2'].append(i2)
                        combined_array['z2'].append(z2)
                        combined_array['y2'].append(y2)

                        combined_array['u'].append(u)
                        combined_array['g'].append(g)
                        combined_array['r'].append(r)
                        combined_array['i'].append(i)
                        combined_array['z'].append(z)
                        combined_array['y'].append(y)
            
        #if ii > 500 :
        #    break
        
#for field in combined_field:
#    print(field)
#    combined_array[field] = np.concatenate(combined_array[field])
    
combined_nparray = np.core.records.fromarrays([combined_array[field] for field in combined_field],\
                                         names=combined_field,\
                                         formats=combined_dtype)
```

```python
from astropy.table import Table

combined_Table = Table(combined_nparray)

combined_Table.write('./01_results/GalacticDWDs.ecsv', overwrite=True)
```
