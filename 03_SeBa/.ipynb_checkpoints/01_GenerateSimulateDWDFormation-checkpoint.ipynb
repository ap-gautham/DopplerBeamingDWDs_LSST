## Star formation history

https://arxiv.org/pdf/1911.02576 - SFR for thin/thick disk

https://arxiv.org/pdf/1905.11404 / https://iopscience.iop.org/article/10.3847/1538-4357/ac8675/pdf- SFR for bulge

### Main Idea
In each time bin, we can calculate the mass of the stars formed. Randomly select N number of stars such that their total mass is equal to the star formation mass.

### Execution
Primary star mass and single star mass is randomly selected. The mass of secondary star in a binary is selected from a uniform distribution. On average, for a large enough sample, the secondary will have half the mass that of primary star. Also, primary star will have same average mass as a single star. Thus the ratio of single star to binary stars is 1/(1+0.5) or 0.66 ( [https://arxiv.org/pdf/2111.13700] - this paper got 0.64]. 

So, if 1000 Msun stars are formed in time t, then 60% of it is assigned to the binaries or 400 Msun is assigned to the primary star. Now the question is how to select 400 Msun by random selection in an efficient manner.

Even if just 0.9-10 Msun is simulated, binaries will be formed along the entire 0.1-150 Msun mass range - 
See Equation (2) : 10.1093/mnras/stz2834

Using Kroupa mass function we find that nearly half the mass is in the range 0.9-10 Msun. 

This results in 200 Msun mass assigned to primary star of the binary in mass range 0.9-10 Msun

Assuming that the total mass is a very large number compared to masses of individual stars (and not 200 Msun), then we can divide the total mass by the expectation value of the Kroupa mass function in the range 0.9-10 Msun to calculate the number of stars to be sampled. So each time, we can sample exactly this number of stars. If this sampling gives total mass much different (beyond, say, 10%) then we can keep sampling with slight variation to the number of stars. 

## Plan

1. $\color{blue}\boxed{\checkmark}$ Generate 3 population of binaries in the range 0.9-10 Msun with 3 different metallicites. In total, $3\times 10^8$ stars are simulated.

    a. $\textbf{Generated $1.2\times 10^7$ stars for each metallicity bin.}$

3. Starting at time t = 13 Gyrs (time of milkyway formation) we simulate star formation. Split the time in $\texttt{np.arange(0,12,0.5)}$. Calculate the star formation in each time bin and estimate the primary mass that is formed in each bin.

4. In each bin, sample N masses (estimated using the expectation value ~ 2.11 M$_\odot$, see below) from the primary mass of simulated stars such that the total mass matches the star formation mass.

5. For each sampled star, using the formation time, we calculate the current age and find the evolution stage of the star at current age (evolution stage at a given time is given by SeBa output).

Thus we can estimate the current population of DWDs.

### 1. Generate $1.2 \times 10^7$ stars with SeBa for each metallicity bin

### 2. Simulate star formation history
- Split $t_{Age}$ = 12 Gyrs into 15 bins : $\texttt{np.arange(0,12,0.5)}$ - even if SFR can high large slope, we only care about total mass of stars formed. We also need to ensure that formation age is well estimated. Hence, linear spacing is necessary and good enough.
- Split into disk/bulge
- Use (R, z) distribution to get a location and assign $\phi \in [0, 2\pi] $ for disk/bulge
- <s>Split total stars into same time bins but with probability</s> Using star formation history, draw DWDs, with replacement, from the simulated population. Assign metallicities as well.
- For each formed star, calculate the remaining time and get the current evolutionary stage.

### 3. Use the total mass of disk/bulge to scale the simulation

Hence we find :
#### Fraction of mass in 0.95-10 Msun range : 0.52
#### Mean mass of stars in 0.95-10 Msun range : 2.11 Msun

# Metallicity

$\mathrm{[Fe/H](R,\tau) = F_m + \nabla[Fe/H] R - (F_m + \nabla[Fe/H] R^{now}_{[Fe/H]=0})f(\tau)}$ 

where, $f(\tau) = \big( 1 - \frac{\tau}{\tau_m} \big)^{\gamma_{[Fe/H]}}$ and $\tau$ is the lookback time

$\mathrm{F_m = -1~dex \quad \nabla[Fe/H] = -0.075~kpc^{-1} \quad R^{now}_{[Fe/H]=0} = 8.7~kpc}\quad \gamma_{[Fe/H]}=0.3$

$\mathrm{Z = Z_{\odot}10^{0.977[Fe/H]}}$

(https://doi.org/10.3847/1538-4357/ac8675)

```python
Z_sun = 0.014
ageMilkyWay = 13.0

F_m = -1.0
grad_FeH = -0.075
R_FeH_now = 8.7
gamma_FeH = 0.3

tau_m = ageMilkyWay

def f_ftau(tau):
    return (1 - tau / tau_m)**gamma_FeH

def f_fe_h(R, tau):
    term_1 = F_m + grad_FeH * R
    term_2 = F_m + grad_FeH * R_FeH_now
    
    fe_h_val = term_1 - (term_2) * f_ftau(tau)
    
    return fe_h_val

def f_Z_metallicity(t, R):
    tau = ageMilkyWay - t
    return Z_sun * 10**(0.977 * f_fe_h(R, tau))
```

# Spatial Distribution

```python
def f_probabilityOfR(R,t,component):
    tau = ageMilkyWay - t

    """
        if component == 'thin_disk':
            R_scale = 3*(1-0.3*(tau/8)) # Wagg et al.
        elif component == 'thick_disk':
            R_scale = 2 # Bland-Hawthorn & Gerhard
        elif component == 'bulge':
            R_scale = 0.5 # Bland-Hawthorn & Gerhard
    """
    
    R_scale_array = np.array([3*(1-0.3*(tau/8)),2,0.5])
    R_scale = R_scale_array[component]
    
    p_R = np.exp(-R/R_scale)*R/R_scale**2 
    
    # Consider a 2D distribution of stars to get this R/R_scale**2
    # Further, for galaxies, the surface brightness/luminosity, L, follows an exponential distribution
    # The number of sources (and thus the probability) is proportional to L
    # N(R) = Integral(rho(r,theta,z)dr) =  N0 * Integral(p(r)dr,[0,R]) 
    # i.e number of sources within radius R is integral of probability density
    # N0*p(R) ~ dN(r)/dr
    # p(R) ~ (L(r)*r*dr)/dr ~ L(r)*r

    return p_R
    
def f_probabilityOfZ(z,component):

    """
    if component == 'thin_disk':
        z_scale = 0.3 # Bland-Hawthorn & Gerhard
    elif component = 'thick_disk':
        z_scale = 0.9 # Bland-Hawthorn & Gerhard
    elif component == 'bulge':
        z_scale = 0.2 # Bland-Hawthorn & Gerhard
    """
    
    z_scale_array = np.array([0.3,0.9,0.2])
    z_scale = z_scale_array[component]
    
    p_z = np.exp(-z/z_scale)/z_scale

    return p_z
    
def f_probabilityOfphi(size):
    """
    phi is defined based on astropy galactocentirc coordinates with earth to the left of galactic center
    phi is drawn from [-np.pi,np.pi]
    """
    p_phi = 1/(2*np.pi)*np.ones(size)
    return p_phi

def f_sample_coords(t,component,size):
    tau = ageMilkyWay - t
    
    R_scale_array = {'thin_disk':3*(1-0.3*(tau/8)),'thick_disk':2,'bulge':0.5}
    R_scale = R_scale_array[component]

    sample_R = scipy.stats.gamma.rvs(a=2,scale=R_scale,size=size)
    sample_R = np.array(sample_R,dtype=np.float16)
    
    z_scale_array = {'thin_disk':0.3,'thick_disk':0.9,'bulge':0.2}
    z_scale = z_scale_array[component]
    
    sample_Z = np.random.choice([1,-1],size=size)*scipy.stats.gamma.rvs(a=1,scale=z_scale,size=size)
    sample_Z = np.array(sample_Z,dtype=np.float16)

    sample_phi = scipy.stats.uniform.rvs(scale=2*np.pi,size=size)
    sample_phi = np.array(sample_phi,dtype=np.float16)

    return sample_R,sample_phi,sample_Z
    
def f_astropyCoords(r,phi,z):
    # Convert (r,phi,z) to astropy coords, which can then be easily converted to (ra,dec) and (l,b)
    # Sun is placed to the left of galactic center

    x = r*np.cos(phi)
    y = r*np.sin(phi)
    z = z
    
    coords = coordinates.SkyCoord(x=x,y=y,z=z,frame=astropy.coordinates.Galactocentric)
    
    return coords
```

# Star Formation Rate

```python
def starFormationRate_thinDisk(t):
    t = ageMilkyWay - t
    return 2.6*(1/(1+(t/9.)**25))*(1+0.3*t/9.)

def starFormationRate_thickDisk(t):
    t = ageMilkyWay - t
    return 9*np.exp(-(t-9.8)**2/2/0.6**2)

def starFormationRate_bulge(t):
    t = ageMilkyWay - t
    return 2.5*np.exp(-(t-8)**2/2/1.5**2)

def starFormationRate_total(t):
    return starFormationRate_thinDisk(t)+\
            starFormationRate_thickDisk(t)+\
                starFormationRate_bulge(t)
```

```python
dt = 0.0001
t = np.arange(0,12,dt)

mass_thick = np.trapz(starFormationRate_thickDisk(t),t,dx=dt)*1e9
mass_thin = np.trapz(starFormationRate_thinDisk(t),t,dx=dt)*1e9
mass_bulge = np.trapz(starFormationRate_bulge(t),t,dx=dt)*1e9

print(f'Thin disk mass = {mass_thin:.2e} Msun')
print(f'Thick disk mass = {mass_thick:.2e} Msun\n')
print(f'Total disk mass = {(mass_thin+mass_thick):.2e} Msun')
print(f'Bulge mass = {mass_bulge:.2e} Msun')

fig,ax = plt.subplots(nrows=1,ncols=2,figsize=(11,5.5),tight_layout=True)

#ax[0].plot(t,full_interpolate(t),'--',c='r',label='Observed')

ax[0].plot(t,starFormationRate_thinDisk(t),'--',c='C0',label='Thin disk')
ax[0].plot(t,starFormationRate_thickDisk(t),'--',c='C2',label='Thick disk')
ax[0].plot(t,starFormationRate_bulge(t),'--',c='C1',label='Bulge')
ax[0].plot(t,starFormationRate_total(t),'-',c='k',label='Combined')

#ax[0].set_xlabel('Lookback Time (Gyrs)')
ax[0].set_xlabel('Age of the Galaxy (Gyrs)')
ax[0].set_ylabel('Star Formation Rate (M$_{\\odot}$/yr)')

ax[1].plot(t,np.cumsum(starFormationRate_thinDisk(t))*(dt)/10,'--',c='C0',label='Thin disk')
ax[1].plot(t,np.cumsum(starFormationRate_thickDisk(t))*(dt)/10,'--',c='C2',label='Thick disk')
ax[1].plot(t,np.cumsum(starFormationRate_bulge(t))*(dt)/10,'--',c='C1',label='Bulge')
ax[1].plot(t,np.cumsum(starFormationRate_total(t))*(dt)/10,'-',c='k')

ax[1].set_xlabel('Age of the Galaxy (Gyrs)')
ax[1].set_ylabel('Total Stellar Mass ($\\times 10^{10}$ M$_{\\odot}$)')

ax[0].legend();
```

```python
toalStarMass = mass_thick+mass_thin+mass_bulge
totalStarsSimulation = toalStarMass*0.4*0.5/2.11

NStarsSimulation = int(np.ceil(totalStarsSimulation)/4)#100*1e6
divideForNMillionStars = round(totalStarsSimulation/NStarsSimulation)
divideForNMillionStars,NStarsSimulation
```

# Start the simulation

```python
kroupaMean_simulation = 2.11

dt_split = 0.1
t_split = np.arange(0,ageMilkyWay+dt_split,dt_split)

temp = 0
ttt = []

tAge_array,R_array,phi_array,z_array,Z_array = [],[],[],[],[]

for ii in tqdm(range(len(t_split)-1)):
    dt_grid = 0.010
    t_grid = np.arange(t_split[ii],t_split[ii+1],dt_grid)

    thickDisk_starMassFormed = np.trapz(starFormationRate_thickDisk(t_grid),t_grid,dx=dt_grid)*1e9
    thinDisk_starMassFormed = np.trapz(starFormationRate_thinDisk(t_grid),t_grid,dx=dt_grid)*1e9
    bulge_starMassFormed = np.trapz(starFormationRate_bulge(t_grid),t_grid,dx=dt_grid)*1e9
    
    t_primaryStarMass = 0.4*thickDisk_starMassFormed    
    t_primaryStarMass_simulation = t_primaryStarMass*0.5    
    thickDisk_numberOfStars = int(np.ceil(t_primaryStarMass_simulation/kroupaMean_simulation/divideForNMillionStars))

    t_primaryStarMass = 0.4*thinDisk_starMassFormed    
    t_primaryStarMass_simulation = t_primaryStarMass*0.5    
    thinDisk_numberOfStars = int(np.ceil(t_primaryStarMass_simulation/kroupaMean_simulation/divideForNMillionStars))

    t_primaryStarMass = 0.4*bulge_starMassFormed    
    t_primaryStarMass_simulation = t_primaryStarMass*0.5    
    bulge_numberOfStars = int(np.ceil(t_primaryStarMass_simulation/kroupaMean_simulation/divideForNMillionStars))

    thick_disk_coords = f_sample_coords(t_split[ii]+dt_split/2,component='thick_disk',size=thickDisk_numberOfStars)
    
    tAge_array.append(np.random.choice(ageMilkyWay-t_grid,size=len(thick_disk_coords[0])))
    R_array.append(thick_disk_coords[0])
    phi_array.append(thick_disk_coords[1])
    z_array.append(thick_disk_coords[2])
    
    thin_disk_coords = f_sample_coords(t_split[ii]+dt_split/2,component='thin_disk',size=thinDisk_numberOfStars)
    
    tAge_array.append(np.random.choice(ageMilkyWay-t_grid,size=len(thin_disk_coords[0])))
    R_array.append(thin_disk_coords[0])
    phi_array.append(thin_disk_coords[1])
    z_array.append(thin_disk_coords[2])
    
    bulge_coords = f_sample_coords(t_split[ii]+dt_split/2,component='bulge',size=bulge_numberOfStars)

    tAge_array.append(np.random.choice(ageMilkyWay-t_grid,size=len(bulge_coords[0])))
    R_array.append(bulge_coords[0])
    phi_array.append(bulge_coords[1])
    z_array.append(bulge_coords[2])

    """
    thick_disk_astropyCoords = \
                    f_astropyCoords(thick_disk_coords[0]*units.kiloparsec,\
                                    thick_disk_coords[1]*units.radian,\
                                    thick_disk_coords[2]*units.kiloparsec)
    thin_disk_astropyCoords = \
                    f_astropyCoords(thin_disk_coords[0]*units.kiloparsec,\
                                    thin_disk_coords[1]*units.radian,\
                                    thin_disk_coords[2]*units.kiloparsec)
    bulge_astropyCoords = \
                    f_astropyCoords(bulge_coords[0]*units.kiloparsec,\
                                    bulge_coords[1]*units.radian,\
                                    bulge_coords[2]*units.kiloparsec)
    """
    
tAge_array = np.array(np.concatenate(tAge_array),dtype=np.float32)
R_array = np.array(np.concatenate(R_array),dtype=np.float16)
phi_array = np.array(np.concatenate(phi_array),dtype=np.float16)
z_array = np.array(np.concatenate(z_array),dtype=np.float16)

coords_array = f_astropyCoords(R_array*units.kiloparsec,phi_array*units.radian,z_array*units.kiloparsec).icrs
print(1)
coords_ra_array = np.array(coords_array.ra.value,dtype=np.float16)
coords_dec_array = np.array(coords_array.dec.value,dtype=np.float16)
coords_distance_array = np.array(coords_array.distance.value,dtype=np.float16)
```

```python
tAge_unique, tAge_counts = np.unique(tAge_array,return_counts=True)

# --- This reduces the number of stars to NStarsSeBa --- #
NStarsSeBa = 50e6
tAge_counts = np.ceil(tAge_counts*(NStarsSeBa/np.sum(tAge_counts)))

numberOfThreads = int(48*2)

# --- This splits the total stars into 96 bins such that each bin has apparoximately equal number of stars --- #
cum_arr = tAge_counts.cumsum() / tAge_counts.sum()
idx = np.searchsorted(cum_arr, np.linspace(0, 1, numberOfThreads, endpoint=False)[1:])
# idx gives locations where the array should be split so that each split has equal number of stars
tAge_counts_split = np.split(tAge_counts, idx)
tAge_unique_split = np.split(tAge_unique, idx)
# --- # --- #

randomSeed = np.random.randint(0,10*len(tAge_array),size=len(tAge_unique))

import os
base_directory = '/home/gadaman1/01_Research/01_Projects/01_WD/08_LSST_DWD/01_SeBa'
os.chdir(base_directory)

countBinaries = 0
countUnique = 0
for ii in range(len(tAge_unique_split)):
    path = base_directory+'/combined_SeBa'+'/'+f'run_{ii:02d}'
    if not os.path.exists(path):
        os.mkdir(path)
    os.system('cp package_SeBa/dstar/SeBa '+path)

    os.chdir(path)

    run_file = [f': > SeBa_output.data\n: > SeBa_input.data']

    for jj in range(len(tAge_unique_split[ii])):
        run_file.append(f'\n./SeBa -R -n {tAge_counts_split[ii][jj]} -N {countBinaries} -m 0.9 -M 10 -z 0.014 '+ \
                          f'-F Kroupa -G Power_Law -y -1 -a 0.1 -U Thermal_Distribution -e 0.01 '+ \
                          f'-P Flat_q -T {(tAge_unique_split[ii][jj]*1000):0.0f} '+ \
                          f'-s {randomSeed[countUnique]} -D')
        
        #run_file.append(f'\n./SeBa -R -n {tAge_counts_split[ii][jj]} -N {countBinaries} -m 0.9 -M 10 -z 0.014 '+ \
        #                  f'-F Kroupa -G Power_Law -y -1 -a 0.1 -U Thermal_Distribution -e 0.01'+ \
        #                  f'-P Flat_q -T {(tAge_unique_split[ii][jj]*1000):0.0f} '+ \
        #                  f'-s {randomSeed[countUnique]} -D')
        
        countBinaries += tAge_counts_split[ii][jj]
        countUnique+=1

    run_file = ''.join(run_file)

    os.system(': > run_binary.sh')
    with open("run_binary.sh", "r+") as f:
        old = f.read() # read everything in the file
        f.seek(0) # rewind
        f.write(run_file) # write the new line before

    os.system('chmod +x run_binary.sh')
    os.chdir(base_directory)
    
# All the coordinates
input_array = np.core.records.fromarrays([np.round(tAge_array*1e3,0),\
                                          R_array,phi_array,z_array,\
                                          coords_ra_array,coords_dec_array,coords_distance_array],\
                                         names=['age',\
                                                'r_cylindrical', 'phi_cylindrical', 'z_cylindrical',\
                                                'RA_ICRS','DEC_ICRS', 'd_ICRS'],\
                                         formats=[np.int32,\
                                                  np.float32,np.float32,np.float32,\
                                                  np.float32,np.float32,np.float32])

#save_directory = './01_results'
#with h5py.File(save_directory+f'/input_array.h5', 'w') as file:
#    file.create_dataset(f'input_array', data=input_array)
    
np.save('01_results/input_array.npy',input_array)

# Unique age
input_unique_array = np.core.records.fromarrays([randomSeed,np.round(tAge_unique*1e3,0),tAge_counts],\
                                         names=['rnd_seed','age_unique','age_unique_counts'],\
                                         formats=[np.int32,np.int32,np.int32])

#save_directory = './01_results'
#with h5py.File(save_directory+f'/input_unique_array.h5', 'w') as file:
#    file.create_dataset(f'input_unique_array', data=input_unique_array)
    
np.save('01_results/input_unique_array.npy',input_unique_array)
```
